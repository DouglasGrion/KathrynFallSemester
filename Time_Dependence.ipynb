```python
def get_MeanStd(data):
    shape = np.shape(data)
    rows = shape[0]
    meanlist = np.zeros(rows)
    stdlist = np.zeros(rows)
    for i in range(rows):
        meanlist[i] = np.nanmean(data[i,:])
        stdlist[i] = np.std(data[i,:])
    return meanlist, stdlist
```
